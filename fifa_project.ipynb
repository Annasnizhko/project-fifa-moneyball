For the overall players, the features related to being a GoalKeeper are not correlated, together with Age, Hight, Weight and GK Rating. Then we remove those features.

```python
numerical_features = ['Overall', 'Potential', 'TotalStats', 'BaseStats',
                      'Skill Moves',
       'International Reputation', 'Pace Total', 'Shooting Total', 'Passing Total',
       'Dribbling Total', 'Defending Total', 'Physicality Total', 'Crossing',
       'Finishing', 'Heading Accuracy', 'Short Passing', 'Volleys',
       'Dribbling', 'Curve', 'Freekick Accuracy', 'LongPassing', 'BallControl',
       'Acceleration', 'Sprint Speed', 'Agility', 'Reactions', 'Balance',
       'Shot Power', 'Jumping', 'Stamina', 'Strength', 'Long Shots',
       'Aggression', 'Interceptions', 'Positioning', 'Vision', 'Penalties',
       'Composure', 'Marking', 'Standing Tackle', 'Sliding Tackle', 'ST Rating',
       'LW Rating', 'LF Rating', 'CF Rating', 'RF Rating', 'RW Rating',
       'CAM Rating', 'LM Rating', 'CM Rating', 'RM Rating', 'LWB Rating',
       'CDM Rating', 'RWB Rating', 'LB Rating', 'CB Rating', 'RB Rating']
categorical_features = ['Attacking Work Rate', 'Defensive Work Rate']
target = 'Value(in Euro)'
```

We assume that we want to sell the 3 oldest players and we manage to sell them at the Value(In Euro) price.
We then have a budget that we want to spend to buy 3 new players, and we divide this budget uniformly between the three.

```python
budget = df[df['Club Name'] == 'Genoa'].sort_values('Age', ascending=False).head(3)[target].sum()
print(f'Estimated budget of future sell {budget}')
budget_per_player = budget/3.
print(f'Estimated budget of each new player {budget_per_player}')
```

# One Hot Encoding of categorical features

```python
encoded_categorical_features = []
for c in categorical_features:
    for v in df[c].unique():
        print(c + '_' + v)
        encoded_categorical_features.append(c + '_' + v)
```

```python
encoded_df = pd.get_dummies(df, columns=categorical_features)
```

```python
encoded_df[encoded_categorical_features + numerical_features]
```

# Player 1

```python
encoded_df[encoded_df['Club Name'] == 'Genoa'].sort_values('Age', ascending=False).head(3)
```

We decide to search for new players at the condition that they are younger than 25 years old  that the declared Valuation is below or equal to our budget, and that the Best Position is the same role we need to substitute. 

```python
young_condition = encoded_df['Age']< 25
```

```python
budget_condition = encoded_df[target] <= budget_per_player
```

```python
player_type_condition = encoded_df['Best Position'] == 'ST'
```

```python
conditions_for_candidates = young_condition & budget_condition & player_type_condition
```

```python
conditions_for_training = ~conditions_for_candidates 
```

```python
encoded_df[conditions_for_candidates]
```

## Train Test split

We decide to use the data of our candidate players as "test set" (in the sense that we will just predict on them) and to train the model of the remaining data

```python
train_data = encoded_df[conditions_for_training]
```

```python
test_data = encoded_df[conditions_for_candidates]
```

```python
test_data.head(5)
```

## Separate features from target

```python
X_train = train_data[numerical_features + encoded_categorical_features]
```

```python
X_train.head(4)
```

```python
X_test = test_data[numerical_features + encoded_categorical_features]
```

```python
Y_train = train_data[target]
```

```python
Y_test = test_data[target]
```

## Normalize features with z-score (make them zero mean and variance 1)

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X_train)
```

```python
normalized_X_train = scaler.transform(X_train)
normalized_X_test = scaler.transform(X_test)
```

## Linear Regression

We decide to use Linear Regression because we do not want to fit too much the data, 
as we believe that the current Value(In Euro) of the players can be not always correct, and hope to use this as our advantage to find actual Value of the candidates players

```python
from sklearn.linear_model import LinearRegression

lr_model = LinearRegression()
```

```python
lr_model.fit(normalized_X_train, Y_train)
```

```python
Y_train.shape
```

```python
Y_test_pred = lr_model.predict(normalized_X_test)
```

```python
delta_predicted_and_market_value = (Y_test_pred - Y_test).astype(int)
```

```python
delta_predicted_and_market_value.name = 'delta_predicted_and_market_value'
```

```python
test_data_with_pred = pd.concat([delta_predicted_and_market_value, test_data], axis=1)
```

```python
test_data_with_pred['Predicted Value'] = Y_test_pred
test_data_with_pred['Predicted Value'] = test_data_with_pred['Predicted Value'].astype(int)
```

```python
info_to_view = ['Full Name', 'delta_predicted_and_market_value', 'Predicted Value', 'Value(in Euro)', 'Overall', 'Potential', 'Age','Best Position','Nationality']
```

```python
test_data_with_pred.sort_values('delta_predicted_and_market_value', ascending=False)[info_to_view]
```

```python
from sklearn.metrics import r2_score
r2_score(Y_test, Y_test_pred)
```

```python
r2_score(Y_train, lr_model.predict(normalized_X_train))
```
